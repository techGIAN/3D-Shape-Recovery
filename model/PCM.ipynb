# **PVCNN**

Authors concoludes in the [paper](https://arxiv.org/pdf/2012.09365.pdf), PVCNN is the network that performs well in the task. The point cloud module depends on Point-Voxel cnn (PVCNN) from the paper [Point-Voxel CNN for Efficient 3D Deep Learning](https://arxiv.org/pdf/1907.03739.pdf) (Liu, Tang, Lin & Han, 2019). 

PVCNN is a hardware-efficient primitive for 3D deep learning. PVCNN combines the advantages of point-based methods and voxel-based methods. A brief explaintation of three important fields of 3D neural networks will be given below:
- **Hardware-Efficient Deep Learning**: Aim to reduce the memory access cost by pruning and quantizing the model.
- **Voxel-Based 3D Models**: Relied on the volumetric representation to process the 3D data.
- **Point-Based 3D Models**: Takes advantage of the symmetric function to process the unordered point sets in 3D.

**PVCNN** is a new hardware-efficient primitive for 3D deep learning, and it takes both advantages from voxel and point based 3D models, minimizes the bottleneck of two methods. The following steps for the feature aggregation are perfromed in the volumetric domain due to its regularity.
- **Normalization**: Normalize the coordinates $\{p_k\}$ before coverting the point clound into the volumetric domain. Firstly, translate all points into the local coordinate system with the gravity center as origin; Secondly, normalize the points into the unit sphere by dividing all coordinates by $\|p_k\|_2$; Thirdly, scale the points to $[0,1]$.
- **Voxelization**
- **Feature Aggregation**
- **Devoxelization**
